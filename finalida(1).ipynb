# Introduction: What is Twitter?

Twitter is an American microblogging and social networking service on which users post and interact with messages known as "tweets". Registered users can post, like and retweet tweets, but unregistered users can only read them. Users access Twitter through its website interface, through Short Message Service (SMS) or its mobile-device application software ("app"). Twitter, Inc. is based in San Francisco, California, and has more than 25 offices around the world. Tweets were originally restricted to 140 characters, but was doubled to 280 for non-CJK languages in November 2017. Audio and video tweets remain limited to 140 seconds for most accounts.

Twitter was created by Jack Dorsey, Noah Glass, Biz Stone, and Evan Williams in March 2006 and launched in July of that year. By 2012, more than 100 million users posted 340 million tweets a day, and the service handled an average of 1.6 billion search queries per day. In 2013, it was one of the ten most-visited websites and has been described as "the SMS of the Internet". As of 2018, Twitter had more than 321 million monthly active users.

# Data Preparation

```python
import pandas as pd
import numpy as np
df = pd.read_csv('tweets.csv')
df[:]
```

```python
data_valid = pd.DataFrame() #checking for null values
data_valid['Columns'] = list(df.columns)
data_valid['Count'] = list(df.count())
data_valid[:]
```

For the puprose of our analysis we don't need columns such as country, latitude and longtitude, we will drop them because of this and also because most of the values are missing. 

```python
df = df.drop(columns = ['country', 'latitude', 'longitude'])
df[:]
```

Language on some of the tweets cannot be identified or just missing at all, since such data entries marked as und, we should substitute them by NaN.

```python
df.replace("und", np.nan, inplace = True)
data_valid = pd.DataFrame() 
data_valid['Columns'] = list(df.columns)
data_valid['Count'] = list(df.count())
data_valid[:]
```

## Data Visualization

### Q1: Find the most liked or shared posts on the user's page.

```python
#creating a dataframe only for most liked posts of a person
df2 = df
df2 = df.sort_values(by=['number_of_likes'], ascending = False).groupby(['author']) #sorting and grouping by author
df2 = df2.first() #and then taking only first values
df2[:]
```

```python
#creating a dataframe only for most shared posts of a person
df3 = df
df3 = df.sort_values(by=['number_of_shares'], ascending = False).groupby(['author']) #sorting and grouping by author
df3 = df3.first() #and then taking only first values
df3[:]
```

```python
df4 = pd.DataFrame()
df4 = pd.merge(df2[['number_of_likes']], df3[['number_of_shares']], on = 'author')
df4[:]
```

```python
axes = df4.plot.bar(figsize = (10,10))
```

As we can see from the figure above, most liked post of the dataframe belongs to Selena Gomez. Most shared post belongs to TheEllenShow.

### Q2: Investigate increase in the number of posts during 2016 year

```python
day = list()
month = list()
year = list()
for i in df['date_time'] :
    day.append(int(str(i[:2])))
    month.append(int(str(i[3:5])))
    year.append(int(str(i[6:10])))
post_data = pd.DataFrame()
post_data['author'] = df['author']
post_data['content'] = df['content']
post_data['day'] = day
post_data['month'] = month
post_data['year'] = year
post_data[:]
```

```python
from2016 = pd.DataFrame()
from2016 = post_data[post_data['year']==2016]
from2016[:]
```

```python
from2016_count = pd.DataFrame()
from2016_count = from2016[['month', 'author']].groupby(['month']).count()
from2016_count = from2016_count.rename(columns = {'author' : 'count'})
from2016_count[:]
```

```python
ax=from2016_count.plot.bar(figsize=(18,10))
```

From the bar plot above we can see that we can't say that during the year 2016 number of posts increased with each month.

### Q3: Measure the correlation between posts' likes and shares.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
x = df['number_of_likes']
y = df['number_of_shares']
plt.scatter(x, y, color = 'pink') 
plt.xlabel('Number of Likes')
plt.ylabel('Number of shares')
plt.show()
```

```python
print(f'Correlation coefficient = {np.round(np.corrcoef(x,y)[0,1], 3)}')
```

So clearly, due the correlation coefficient there is pretty high correlation between the number of likes and number of shares of the post

### Q4. Find the most common languages among the posts.

```python
lang = pd.DataFrame()
lang = df[['language', 'content']].groupby(['language']).count()
lang.reset_index(inplace=True)
lang[:]
```

Clearly English is dominant, Spanish is second dominant, so lets consider all other languages except English and Spanish.

```python
lang = lang[lang['language']!='en']
lang = lang[lang['language']!='es']
plt.figure(figsize=(15,15))
plt.pie(lang['content'],  labels=lang['language'], autopct='%1.1f%%',
        shadow=True)
plt.axis('equal') 
plt.show()
```

So English and Spanish are most frequently used languages on twitter. Surprisingly among rest of the languages Portuguese is the most frequent one.

### Q5: Find the most frequent words among the posts.

```python
from collections import Counter
Counter(" ".join(df["content"]).split()).most_common(200)
```

```python
word_list = list(df["content"].apply(lambda x: x.split()))
word_list = [x for y in word_list for x in y]
from wordcloud import WordCloud, STOPWORDS
stopwords = [':', '&amp;', '-', ':)', '1', '2', '3', '4', '5', '6', '7', '8', '9', '0', 'http', 'https']
wordcloud = WordCloud(width = 1200, height = 600, 
                          stopwords=stopwords, background_color='white',
                          max_font_size=100,).generate(" ".join(word_list))
plt.figure(figsize = (15,10))
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis("off") 
plt.show()
```

Quite frequently we can see how President Obama was mentioned, phrases like 'Thank you' and similar other phrases
